```python
from splinter import Browser
browser = Browser('chrome')
from bs4 import BeautifulSoup
import pandas as pd
import time
```

## Scraping the data

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://onlinelibrary.wiley.com/doi/full/10.1111/joss.12237'
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
#results=soup.find_all('div', class_ ='article-table-content')['id'=='joss12237-tbl-0003']
results=soup.find_all('div', class_ ='article-table-content')[2]

print(results)
        
              
```

```python
attrs = []
attributes = results. find_all('td', class_='right-bordered-cell left-aligned')
for attribute in attributes:
    attr= attribute.get_text().strip()
    attrs.append(attr)
print(attrs)
```

```python
defs = []
definitions = results.find_all('td', class_='left-aligned')
for definition in definitions:
    defi = definition.get_text().strip()
    defs.append(defi)
print(defs)
```

```python
attributes =defs[::3]
attributes
```

```python
new_attr=[i.replace('a\nabAttributes that do not have a superscript are first level attributes, those with a superscipt \u201ca\u201d are second level attributes underneath a first level (e.g., citrus underneath fruity), and those with a superscript \u201cb\u201d are third level attributes underneath a second level (e.g., orange undeath citrus).','') for i in attributes]
new_attr
```

```python
new_attrn=[i.replace('b\nabAttributes that do not have a superscript are first level attributes, those with a superscipt \u201ca\u201d are second level attributes underneath a first level (e.g., citrus underneath fruity), and those with a superscript \u201cb\u201d are third level attributes underneath a second level (e.g., orange undeath citrus).','') for i in new_attr]
new_attrn
```

```python
attr_1=[i.replace('Spices/','') for i in new_attrn]
attr_2=[i.replace('Spice Brown/','') for i in attr_1]
attr_3=[i.replace('Sweet Aromatics/','') for i in attr_2]
attr_4=[i.replace('Brown Sugar/','') for i in attr_3]
attr_5=[i.replace('Nutty/','') for i in attr_4]
attr_6=[i.replace('Chocolate/','') for i in attr_5]
attr_7=[i.replace('Floral/','') for i in attr_6]
attr_8=[i.replace('Fruity/','') for i in attr_7]
attr_9=[i.replace('Dried Fruit/','') for i in attr_8]
attr_10=[i.replace('Citrus Fruit/','') for i in attr_9]
attr_11=[i.replace('Berry/','') for i in attr_10]
attr_12=[i.replace('Stone Fruit/','') for i in attr_11]
attr_13=[i.replace('Fruity, Tropical/','') for i in attr_12]
attr_14=[i.replace('Green/','') for i in attr_13]
attr_15=[i.replace('Tobacco/','') for i in attr_14]
attr_16=[i.replace('Animal/','') for i in attr_15]
attr_17=[i.replace('Musty/','') for i in attr_16]
attr_18=[i.replace('Chemical/','') for i in attr_17]
attr_19=[i.replace('Distilled‐Boozy/','') for i in attr_18]
final_attr=[i.replace('Balance/','') for i in attr_19]
final_attr
```

```python
meanings=defs[1::3]
meanings
```

```python
i_meanings=[i.replace('\u2010','') for i in meanings]
i_meanings
```
